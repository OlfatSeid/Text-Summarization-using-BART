### Extract a Text from the PDF

```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

pdf_path = '/content/MachineLearning-Lecture01.pdf'
document_text = extract_text_from_pdf(pdf_path)
```

## # Embed the Text

```python

embedder = SentenceTransformer('all-MiniLM-L6-v2')
chunks = document_text.split('\n')
embeddings = embedder.encode(chunks, convert_to_tensor=True)
```

## * Retrieve relevant chunks

```python

def retrieve_relevant_chunks(query, chunks, embeddings, top_k=5):
    query_embedding = embedder.encode(query, convert_to_tensor=True)
    cosine_scores = torch.nn.functional.cosine_similarity(query_embedding, embeddings, dim=-1)
    top_k_indices = torch.topk(cosine_scores, top_k).indices
    retrieved_chunks = [chunks[idx] for idx in top_k_indices]
    return retrieved_chunks
```

```python
query = "summary"
retrieved_chunks = retrieve_relevant_chunks(query, chunks, embeddings)

# Prepare data for fine-tuning
tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')
model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')
inputs = tokenizer(retrieved_chunks, truncation=True, padding='max_length', max_length=512, return_tensors='pt')
dataset = Dataset.from_dict({'input_ids': inputs['input_ids'], 'attention_mask': inputs['attention_mask'], 'labels': inputs['input_ids']})
```

## * Fine-Tune the BART Model

```python

training_args = TrainingArguments(
      output_dir='./results',
    num_train_epochs=80,
    per_device_train_batch_size=2,
    learning_rate=1e-6,
    save_steps=20,
    save_total_limit=8,
    logging_dir='./logs',
    logging_steps=10,
    weight_decay=0.01,  # Add weight decay for regularization
    warmup_steps=50,  # Add learning rate warmup
    max_grad_norm=1.0,  # Gradient clipping

)
trainer = Trainer(
    model=model,
    args=training_args,

    train_dataset=dataset,
)
trainer.train()

model_save_path = "fine_tuned_bart"
model.save_pretrained(model_save_path)
tokenizer.save_pretrained(model_save_path)
print(f"Model and tokenizer saved as {model_save_path}")
```

```python
# Load the fine-tuned model and tokenizer
fine_tuned_model = BartForConditionalGeneration.from_pretrained("fine_tuned_bart")
fine_tuned_tokenizer = BartTokenizer.from_pretrained("fine_tuned_bart")

inputs = fine_tuned_tokenizer([document_text], max_length=1024, return_tensors='pt', truncation=True)
summary_ids = fine_tuned_model.generate(
    inputs['input_ids'],
    num_beams=4,
    max_length=1024,
    min_length=512,
    length_penalty=0.6,
    early_stopping=True,
    no_repeat_ngram_size=3,
    forced_bos_token_id=0,
    forced_eos_token_id=2,

)
full_summary = fine_tuned_tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print("Summary:", full_summary)
```

```python
from nltk.translate.bleu_score import sentence_bleu
from rouge_score import rouge_scorer
import textstat

reference_summary = reference_summary
generated_summary =full_summary

bleu_score = sentence_bleu([reference_summary.split()], generated_summary.split())

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
scores = scorer.score(reference_summary, generated_summary)
rouge1_score = scores['rouge1'].fmeasure
rouge2_score = scores['rouge2'].fmeasure
rougeL_score = scores['rougeL'].fmeasure

print(f"BLEU Score: {bleu_score}")
print(f"ROUGE-1 Score: {rouge1_score}") #ROUGE (Recall-Oriented Understudy for Gisting Evaluation)
print(f"ROUGE-2 Score: {rouge2_score}")
print(f"ROUGE-L Score: {rougeL_score}")

# Calculate readability metrics
reading_ease_score = textstat.flesch_reading_ease(generated_summary)
fog_index_score = textstat.gunning_fog(generated_summary)

print(f"Readability (Flesch Reading Ease): {reading_ease_score}")
print(f"Readability (Gunning Fog Index): {fog_index_score}")

```

```python
# Perplexity Score
from transformers import GPT2LMHeadModel, GPT2Tokenizer

model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

def calculate_perplexity(text):
    encodings = tokenizer(text, return_tensors='pt')
    max_length = model.config.n_positions
    stride = 512

    nlls = []
    for i in range(0, encodings.input_ids.size(1), stride):
        begin_loc = max(i + stride - max_length, 0)
        end_loc = min(i + stride, encodings.input_ids.size(1))
        trg_len = end_loc - i
        input_ids = encodings.input_ids[:, begin_loc:end_loc].to(model.device)
        target_ids = input_ids.clone()
        target_ids[:, :-trg_len] = -100

        with torch.no_grad():
            outputs = model(input_ids, labels=target_ids)
            neg_log_likelihood = outputs.loss * trg_len

        nlls.append(neg_log_likelihood)

    ppl = torch.exp(torch.stack(nlls).sum() / end_loc)
    return ppl.item()

perplexity_score = calculate_perplexity(full_summary)
print("Perplexity Score:", perplexity_score)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
original_text="/content/MachineLearning-Lecture01.pdf"
def cosine_similarity_evaluation(original_text,full_summary):
    documents = [original_text, full_summary]
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(documents)
    cosine_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
    return cosine_sim[0][0]

cosine_sim_score = cosine_similarity_evaluation(original_text, full_summary)
print("Cosine Similarity Score:", cosine_sim_score)
```

## URL Text Summarizer with the Fine-Tuned BART Model.

```python
import requests
from IPython.display import display
from transformers import BartTokenizer, BartForConditionalGeneration
from ipywidgets import FileUpload,Text, Textarea, Button, Output, HBox

model_name = "fine_tuned_bart"
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)


url_input = Text(placeholder='Enter URL of the webpage...')

button_fetch_summarize = Button(description="Fetch & Summarize", button_style='success', icon="link")
button_clear_url = Button(description="Clear URL", button_style='danger', icon="trash")

output = Output()

def fetch_text_from_url(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.text
    except requests.exceptions.RequestException as e:
        return str(e)

def summarize_text(text):
    try:
        inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=1024, truncation=True)
        summary_ids = model.generate(inputs, max_length=512, num_beams=4, early_stopping=True)
        summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
        return summary
    except Exception as e:
        return str(e)

def handle_fetch_summarize(b):
    output.clear_output()
    url = url_input.value.strip()
    if url:
        text = fetch_text_from_url(url)
        if isinstance(text, str):
            with output:
                print(f"Error fetching URL: {text}")
        else:
            summary = summarize_text(text)
            with output:
                print("Summary:", summary)
    else:
        with output:
            print("Please enter a valid URL.")

def handle_clear_url(b):
    url_input.value = ''
    output.clear_output()
    with output:
        print("URL cleared.")

button_fetch_summarize.on_click(handle_fetch_summarize)
button_clear_url.on_click(handle_clear_url)

buttons = HBox([button_fetch_summarize, button_clear_url])

display(url_input, buttons, output)
```
